####Data Augmentation using Keras

```
import keras
from keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img

#ImageDataGenerator library helps us to apply all the Data Augmentation properties
datagen= ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```
#Load the image
img=load_img('/content/cat or dog.jpg')
```

```
#View loaded image
img
```

```
#Convert the image into array
#This is a Numpy array with shape (195, 306, 3)
x=img_to_array(img)
```

```
#View the shape of array
x.shape
```

```
#Reshape the array to (1,3,150,150)
x=x.reshape((1,)+x.shape)
```

```
#Check the shape of x
x.shape
```

```
#Datagen has an important function called .flow() command
#.flow() command generates batches of randomly transformed images
#and saves the results to '/content/preview' directory
#We are going to generate only 20 images and save this to /content/preview folder-
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir='/content/preview',save_prefix='cat',save_format='jpeg'):
  i+=1
  if i>20:
    break #If we wont use break, the loop will run indefinitely
```
